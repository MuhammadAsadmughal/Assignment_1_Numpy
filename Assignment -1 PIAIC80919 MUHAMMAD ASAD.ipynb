# NP Random


```python
# np.random.randint

# toss a coin
y = np.random.randint(5)
print(y)

# toss a coin 3 times
a = np.random.randint(5,size=3)
print(a)

# roll a dice
b = np.random.randint(2,8)
print(b)

# roll a dice 10 times
x = np.random.randint(2,8,size=3)
print(x)
```

```python
# toss a coin
f = np.random.choice(["Head","Tail"])
print(f)

# toss a coin 3 times
g = np.random.choice(["Head","Tail"],size=3)
print(g)

# roll a dice
h = np.random.choice([1,2,3,4,5,6])
print(h)

# roll a dice 3 times
i = np.random.choice([1,2,3,4,5,6],size=3)
print(i)
```

```python
# np.random.binomial

# number of successes (number of heads) in 5 coin flips
x = np.random.binomial(6,0.3)
print(x)

# number of successes (number of heads) in 5 coin flips - biased coin with a 0.7 probability of obtaining heads
y = np.random.binomial(6,0.7)
print(y)

# we can obtain approximated probabilities by simulating a huge number of flips probability of obtaining 4 head in 5 flips 
flips = np.random.binomial(6,0.5,size=int(1e6))
(flips==4).mean()
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
print(s)

#Verify the mean and the variance:
a = abs(sigma - np.std(s, ddof=1)) < 0.01
b = abs(mu - np.mean(s)) < 0.01
print(a,b)
```

```python
# np.random.poisson and then graph using matplotlib

s = np.random.poisson(20, 10000)
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 12, density=True)
plt.show()
```

```python
# np.random.power()
x = 5. # shape
samples = 1000
s = np.random.power(x, samples)
count, bins, ignored = plt.hist(s, bins=30)
y = np.linspace(0, 2, 100)
z = x*y**(x-1.)
normed_y = samples*np.diff(bins)[0]*z
plt.plot(y, normed_y)
plt.show()
```

```python
# np.random.uniform()

s = np.random.uniform(-1,0,1000)
count, bins, ignored = plt.hist(s, 20, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=3, color='r')
plt.show()
```

```python
#numpy.ndarray.byteswap()
a = np.array([1, 360, 9546], dtype = np.int16) 

print ('Our array is: ', a)   

print ('Representation of data in memory in hexadecimal form: ', map(hex,a))  

# byteswap() function swaps in place by passing True parameter 

print ('Applying byteswap() function: ',a.byteswap(True)) 

print ('In hexadecimal form: ',map(hex,a))
```

```python
#numpy.sort()
a = np.array([[5,9],[8,2]]) 

print ('Our array is: ',a,'\n') 

print ('Applying sort() function: ',np.sort(a),'\n') 
  
print ('Sort along axis 0:') 
print (np.sort(a, axis = 0)) 
print ('\n')  

# Order parameter in sort function 
dt = np.dtype([('name', 'S10'),('age', int)]) 
a = np.array([("ali",25),("robert",28),("mani", 19), ("amir",37)], dtype = dt) 

print ('Our array is: ',a,'\n') 

print ('Order by name:') 
print (np.sort(a, order = 'name'))
```

```python
#numpy.argsort()
x = np.array([6,4,9,3,5, 1, 2]) 

print ('Our array is: ',x,'\n') 

print ('Applying argsort() to x: ') 
y = np.argsort(x) 
print (y,'\n') 

print ('Reconstruct original array in sorted order:',x[y],'\n' ) 

print ('Reconstruct the original array using loop:') 
for i in y: 
   print (x[i])
```

```python
#numpy.lexsort()
nm = ('ahmed','sunny','rauf','amir') 
dv = ('a.y.', 'c.y.', 'c.y.', 'f.y.') 
ind = np.lexsort((dv,nm)) 

print ('Applying lexsort() function:',ind,'\n') 

print ('Use this index to get sorted data:') 
print ([nm[i] + ", " + dv[i] for i in ind]) 
```

```python
#numpy.nonzero()
a = np.array([[30,40,0],[0,20,10],[50,0,60]]) 

print ('Our array is:','\n') 

print ('Applying nonzero() function:') 
print (np.nonzero (a))
```

```python
#numpy.where()
a = np.arange(8.).reshape(2, 4) 

print ('Our array is: ',a) 

print ('Indices of elements > 3') 
y = np.where(a > 3) 
print (y)  

print ('Use these indices to get elements satisfying the condition') 
print (a[y])
```

```python
#numpy.extract()
x = np.arange(9.).reshape(3, 3) 

print ('Our array is: ',x)

# define a condition 
condition = np.mod(x,2) == 0 

print ('Element-wise value of condition',condition) 

print ('Extract elements using condition') 
print (np.extract(condition, x))
```
